```python
articles = []

with open('./data/preprocessed/final_preprocessed_text.txt', 'r', encoding='utf-8') as f:
    total_articles = f.read()
```

```python
raw_articles = total_articles.split('"')
```

```python
raw_articles = raw_articles[1:]
articles = [article.split('\n') for article in raw_articles]
```

```python
print(articles[0])
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=articles,
    vector_size=100,
    window=5,           # 앞, 뒤 5개의 단어와의 연관성을 보겠다
    min_count=5,
    sg=0                # 0: CBOW | 1: Skip-gram
)
```

```python
model.wv.vectors.shape
```
